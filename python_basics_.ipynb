## Data Types
- int:
    What does it store?

    Whole numbers (no decimals)

    Positive or negative

    Examples: 5, -10, 120000

    Maximum size?

    Python integers have no fixed maximum size.
    They grow as large as your RAM can handle.
---------------------------------------------------------
- float
    - What does it store?

    Decimal numbers (numbers with a fractional part)

    Examples: 3.14, 0.5, -2.75

    - Maximum size?

    Python floats follow IEEE 754 double precision:

    Uses 64 bits

    Precision: about 15–17 decimal digits

    - Max value: 1.8 × 10^308

    - Smallest positive value: 5 × 10^-324

    If a number is too large → it becomes inf.
--------------------------------------------------------
- str
    - What does it store?

    Text (letters, digits, symbols)

    - Maximum size?

    No fixed limit.
    A string can be as large as your memory allows.
---------------------------------------------------------
- bool
    - What does it store?

    Only two values:

    True

    False

    - Maximum size?

    Only 1 bit logically, but Python stores it as a small internal object. Size is tiny and not important.


```python
# examples
a = 5
b = 3.14
c = 'Hello'
d = True
print(type(a), type(b), type(c), type(d))
```

## Type Casting

Type casting means converting a variable from one data type to another.

الـ Type Casting يعني تحويل المتغير من نوع بيانات لآخر.
بايثون توفر دوال مثل: int(), float(), str(), bool().

## Operators
Operators are symbols that perform mathematical, logical, or comparison operations.
الـ Operators هي رموز تقوم بـ عمليات حسابية، منطقية، أو مقارنة.

```python
# Arithmetic
a = 10 + 5
print(a)
b = 10 * 2
print(b , "\n")

# Comparison
print(10 > 5)

# Logical
print(a > b and b < a)
```

## Lists
Lists are ordered collections of items.

 الليست هي مجموعة عناصر بترتيب.

```python
my_list = [1, 2, 3]
print(my_list)
print(my_list[0])
```

## List Indexing

Lists are ordered, so you can access elements using indices (starting from 0).
You can also use negative indices to start from the end

الـ list
 مرتبة، فيمكنك الوصول للعناصر باستخدام المؤشرات (تبدأ من 0).
تقدر تستخدم مؤشرات سلبية للوصول من آخر القائمة.

```python
my_list = [10, 20, 30, 40]
print(my_list[0])
print(my_list[-1])
```

## Tuple

 tuple is a collection of items which is ordered and immutable.
 This means you can store multiple values in it, but you cannot change the values after creating it.

 الـ tuple هو مجموعة عناصر مرتبة وغير قابلة للتغيير.
 ده معناه إنك تقدر تخزن عدة قيم جواها، لكن مش هتقدر تعدل القيم بعد ما تنشئ الـ tuple.

```python
my_tuple = (1, 2, 3, 4)
print(my_tuple)
```

## Tuple Indexing

Tuples are ordered just like lists, so you can access elements the same way.
But tuples are immutable, so you cannot change their elements.

الـ tuple مرتبة زي الـ list، فممكن توصل للعناصر بنفس الطريقة.
لكن الـ tuple غير قابل للتغيير، فلا يمكنك تعديل العناصر.

```python
my_tuple = (10, 20, 30, 40)
print(my_tuple[2])
print(my_tuple[-2])
```

## Dictionaries
Dictionaries store data as key-value pairs.

 القاموس بيخزن البيانات بنظام مفتاح وقيمة.

```python
person = {'name': 'Khaled', 'age': 22}
print(person['name'])
```

## Dictionary Indexing

Dictionaries are unordered collections of key-value pairs.
You cannot access elements by index number; you must use the key.

الـ dictionary غير مرتبة، وتتكون من مفتاح-قيمة.
لا يمكن الوصول للعناصر بالرقم، لازم تستخدم المفتاح.

```python
my_dict = {"name": "Khaled", "age": 20}
print(my_dict["name"]) 
print(my_dict["age"])
```

## Conditionals
Conditionals let your program make decisions based on conditions using if, elif, and else.

الجمل الشرطية تسمح للبرنامج بـ اتخاذ قرارات بناءً على شروط باستخدام if, elif, else.

```python
age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

## Loops
Loops allow you to repeat code multiple times. Main types: for and while.

الحلقات التكرارية تسمح لك بـ تنفيذ كود أكثر من مرة. الأنواع الرئيسية: for و while.

```python
# for loop
print("For Loop")
for i in range(3):
    print(i)

print("-"*30)

# while loop
print("While Loop")
count = 0
while count < 3:
    print(count)
    count += 1
```

## Strings
Strings are sequences of characters. Python provides many methods to manipulate strings.

النصوص هي تسلسل من الأحرف. بايثون توفر دوال عديدة للتعامل مع النصوص.

```python
text = "Hello, Khaled"
print(text)  # Hello, Khaled
```

## upper() and lower()
upper() = Converts all characters in a string to uppercase.

lower() = Converts all characters in a string to lowercase.

```python
text = "hello"
print(text.upper())
text = "KHALED"
print(text.lower()) 
```

## strip()

Removes spaces (or other characters) from the beginning and end of the string.

يحذف المسافات (أو حروف محددة) من البداية والنهاية.

```python
text = "  hello  "
print(text.strip())
```

## replace(old, new)

Replaces a substring with another substring.

يستبدل جزء من النص بجزء آخر.

```python
text = "hello world"
print(text.replace("hello", "hi"))
```

## split(separator)

Splits the string into a list based on a separator.

يقسم النص إلى قائمة بناءً على فاصل محدد.

```python
text = "a,b,c"
print(text.split(","))
```

## Functions
Functions are reusable blocks of code.
Functions can take inputs called parameters or arguments.
Parameters are variables defined in the function, arguments are the values you pass

 أكواد بتتكرر فبنكتبها مرة ونستخدمها كتير.
 
 الدوال ممكن تاخد مدخلات تسمى Parameters أو Arguments.

Parameter: المتغير اللي معرف جوه الدالة

Argument: القيمة اللي بتمررها للدالة

```python
def hello_func(name):
    return f'Hello {name}'

print(hello_func('Khaled'))
```

## Return Value.

Functions can return a value using return.
This allows you to use the result outside the function.
--------
الدوال ممكن ترجع قيمة باستخدام return.
ده يسمحلك باستخدام النتيجة خارج الدالة.

```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```
